```python
for i in range(len(filename_list)):
  Okt = konlpy.tag.Okt()
  globals()['Okt_morphs_{}'.format(i)] = Okt.pos(globals()['contents_{}'.format(i)])  # 튜플반환
  print(globals()['Okt_morphs_{}'.format(i)])
```

```python
for i in range(len(filename_list)):
  globals()['NAVN_words_{}'.format(i)] = []
  for word, pos in globals()['Okt_morphs_{}'.format(i)]:
      if pos in ['Noun']: 
          globals()['NAVN_words_{}'.format(i)].append(word)
  print(globals()['NAVN_words_{}'.format(i)])
```

```python
for i in range(len(filename_list)):
  globals()['contents_preprocessed_{}'.format(i)] = " ".join(globals()['NAVN_words_{}'.format(i)])
  print(globals()['contents_preprocessed_{}'.format(i)])
```

```python
from keybert import KeyBERT

kw_model = KeyBERT('distilbert-base-nli-mean-tokens')

for i in range(len(filename_list)):
  globals()['keywords_{}'.format(i)] = kw_model.extract_keywords(globals()['contents_preprocessed_{}'.format(i)])
```

```python
kw_model_2 = KeyBERT()

for i in range(len(filename_list)):
  globals()['keywords2_{}'.format(i)] = kw_model_2.extract_keywords(globals()['contents_preprocessed_{}'.format(i)])
```
